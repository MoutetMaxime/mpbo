```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from BO import BayesianOptimizer
from ObjectiveFunction import ObjectiveFunction
```

**Ackley:**

```python
# Define the objective function
obj = ObjectiveFunction("Ackley", dim=2)
search_space = obj.create_input_space()
response = obj.generate_true_response(search_space)

# Create the Bayesian Optimizer
opt = BayesianOptimizer(search_space, response)


kappa_list = [2, 2.5, 3, 3.5, 4, 5]

last_regret = []
last_exploit_regret = []
for kappa in kappa_list:
    print(f"Kappa: {kappa}")
    regret, _, exploit_regret = opt.train(
        kappa,
        initial_points=1,
        repetitions=30,
        iterations=60,
        training_iter=10,
        strategy="Vanilla BO",
        return_exploit=True,
    )
    last_regret.append(regret.mean(axis=0)[-1])
    last_exploit_regret.append(exploit_regret.mean(axis=0)[-1])
```

```python
plt.plot(kappa_list, last_regret, marker="o", label="Regret")
plt.plot(kappa_list, last_exploit_regret, marker="o", label="Exploit Regret")
plt.legend()
```
